<h2>Process reduce vehicle miles travel</h2>

```python
df = pd.read_excel("VMT_2010_2018.xlsx")
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
year_list = df.columns[2:]
year_list
```

<h2>Region</h2>

```python
pop_region = pop.loc[pop['COUNTY'] =='15-County Region']
pop_region
```

```python
df_region = df.loc[df['COUNTY_NAME'] == 'Region']
df_region
```

```python
#year_list = df_region.columns[2:]
region_result = df_region.iloc[:,2:]*1000*365/pop_region.loc[:,year_list]
```

```python
region_result
```

```python
region= region_result.T.reset_index()
region.columns = ["year", "VMT_per_cap_year"]
region
```

<h2>County</h2>

```python
df_county = df.iloc[:-1].copy()
```

```python

df_county['GEOID'] = df_county['GEOID'].fillna(0)
df_county['GEOID'] = df_county['GEOID'].astype(int).astype(str)
df_county['GEOID'] = df_county['GEOID'].apply(lambda x: x.zfill(3))
df_county['Id2'] = '39'+df_county['GEOID']
df_county.head()
```

```python
county_final = pd.DataFrame()
for year in year_list:
    county_data = df_county.loc[:,[year,'Id2']]
#Index(['2010_x', 'id', '2010_y', 'Id2'], dtype='object')
    county_pop = county_data.merge(pop[[year,'Id2']],on='Id2')
    county_pop['annual'] = county_pop[str(year)+'_x']*1000*365/county_pop[str(year)+'_y']
    county_pop['year'] = year 
    county = county_pop[['Id2','year','annual']]
    county_final = pd.concat([county_final,county],ignore_index=True)
#county_final.reset_index()
    
```
