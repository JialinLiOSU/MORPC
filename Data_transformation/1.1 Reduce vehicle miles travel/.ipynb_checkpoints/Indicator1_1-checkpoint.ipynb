```python
def indicator1_1(pop,year, df):
    pop1 = pop[['Id2', 'COUNTY', year]]
    pop1['county_join'] = pop1['COUNTY'].str[:3]
    pop1['county_join'].replace('MOR', 'MRW', inplace=True)
    df1 = df[['county','total']]
    select_county = df1[df1['county'].isin(['MAR','LOG', 'UNI', 'MAD', 'FAY', 'DEL', 'MRW', 'FRA','PIC', 'ROS', 'KNO', 'LIC', 'FAI', 'PER', 'HOC'])]
    #join pop and select_county
    county_pop = pop1.merge(select_county,left_on = 'county_join',right_on='county')
    county_pop['total']  = county_pop['total'].str.replace(',','')
    county_pop['total'] = county_pop['total'].astype(float)
    county_pop['annual'] = county_pop['total']*1000*365/county_pop[year]
    region = sum(county_pop['total'])*1000*365/sum(county_pop[year])
    final_county = county_pop[['Id2','annual']]
    final_county['year'] = year
    return final_county,region
    
```
