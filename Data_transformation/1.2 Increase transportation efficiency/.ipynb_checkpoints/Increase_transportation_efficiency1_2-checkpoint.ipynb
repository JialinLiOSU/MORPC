<p>Manual Process: filter with interest county for B08301, available datascoure: ACS 3-year for 2008, ACS 5-year (2009-2018),got zip file. In Year 2008, available's county data: Delaware County, Fairfield County, Franklin County, Licking County, Ross County </p>


```python
import requests
import pandas as pd
import numpy as np
```

<h1>Retreive Data using API </h1>

```python
#define function to retreive data
#url2009 = 'https://api.census.gov/data/2009/acs5?&get=GEOID,B08301_001E,B08301_003E,B08301_004E,B08301_010E,B08301_018E,B08301_019E,B08301_021E&for=county:*&in=state:39'
def getdata(year):
    url = 'https://api.census.gov/data/{}/acs/acs5?get=GEO_ID,B08301_001E,B08301_003E,B08301_004E,B08301_010E,B08301_018E,B08301_019E,B08301_021E&for=county:*&in=state:39'.format(str(year))
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()
    df = pd.DataFrame(data[1:], columns = data[0])
    #df['year'] = year
    return df
```

```python
#get 2009 acs
url2009 = 'https://api.census.gov/data/2009/acs5?&get=GEOID,B08301_001E,B08301_003E,B08301_004E,B08301_010E,B08301_018E,B08301_019E,B08301_021E&for=county:*&in=state:39'
response = requests.get(url2009)
response.raise_for_status()
data = response.json()
df2009 = pd.DataFrame(data[1:], columns = data[0])
#df2009['year'] = 2009
df2009.head()
```

```python
#retreive data from 2010 to 2017
df2010 = getdata(2010)
df2011 = getdata(2011)
df2012 = getdata(2012)
df2013 = getdata(2013)
df2014 = getdata(2014)
df2015 = getdata(2015)
df2016 = getdata(2016)
df2017 = getdata(2017)

```

<h1>Process Data</h1>
<p>need year column, select central ohio, group into one df,divide each mode by total workers</p>

```python
df2010.dtypes
```

```python
dfs = [df2009,df2010,df2011,df2012,df2013,df2014,df2015,df2016,df2017]
i = 0
newDF = pd.DataFrame()
for df in dfs:
    df['year'] = 2009+i
    i+=1
    df['id'] = df['state']+df['county']
    df.set_index(['year', 'id'])
    test = df[df['id'].isin(['39101','39091', '39159', '39097', '39047', '39041','39117', '39049','39129','39141','39083', '39089','39045', '39127', '39073'])]
    newDF = newDF.append(test, ignore_index = True)
```

```python
#select data

selectdf = newDF[['B08301_001E','B08301_003E','B08301_004E','B08301_010E','B08301_018E','B08301_019E','B08301_021E','id','year']]
selectdf.rename(columns={'B08301_001E':'Tot',
                        'B08301_003E':'Drove',
                         'B08301_004E':'Carpool',
                         'B08301_010E':'Pub_Transic',
                         'B08301_018E':'Bicycle',
                         'B08301_019E':'Walk',
                         'B08301_021E':'Work_Home'}, inplace=True)
selectdf['Tot']= selectdf['Tot'].astype('float')
selectdf[ 'Drove']= selectdf['Drove'].astype('float')
selectdf['Carpool']= selectdf['Carpool'].astype('float')
selectdf['Pub_Transic']= selectdf['Pub_Transic'].astype('float')
selectdf['Bicycle']= selectdf['Bicycle'].astype('float')
selectdf['Walk']= selectdf['Walk'].astype('float')
selectdf['Work_Home']= selectdf['Work_Home'].astype('float')
selectdf.dtypes
selectdf.head()
```

<h1>Organize data into defined format</h1>

```python
final_year = selectdf.groupby(['year']).sum()
final_year
```

```python
final_year['Drove']  = final_year['Drove']/final_year['Tot']
final_year['Carpool'] =final_year['Carpool']/final_year['Tot']
final_year['Pub_Transic']=final_year['Pub_Transic']/final_year['Tot']
final_year['Bicycle']=final_year['Bicycle']/final_year['Tot']
final_year['Walk']=final_year['Walk']/final_year['Tot']
final_year['Work_Home']= final_year['Work_Home']/final_year['Tot']
final_year.head()
```
