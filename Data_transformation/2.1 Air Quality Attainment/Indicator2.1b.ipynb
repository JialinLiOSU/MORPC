```python
df_require = df_select.loc[df_select['id'].isin(['39101','39091', '39159', '39097', '39047', '39041','39117', '39049','39129','39141','39083', '39089','39045', '39127', '39073'])]
print(len(df_require))
df_require['id'].unique()
```

```python
df_require1 = df_require.loc[df_require['first_max_value'] > 35.4]
print(len(df_require1))
```

```python
df_require1.loc[(df_require1['first_max_value'] >35.4) & (df_require1['first_max_value'] <= 55.4), 'aqi_category'] = "Unhealthy for Sensitive Groups"
df_require1.loc[(df_require1['first_max_value'] >55.4) & (df_require1['first_max_value'] <= 150.4), 'aqi_category'] = "Unhealthy"
df_require1.loc[df_require1['first_max_value'] >150.4 , 'aqi_category'] = "Very Unhealthy"

df_require1.head()
```

```python
result = df_require1.drop_duplicates(subset=['id','date_local'], keep="last")
print(len(result))
```

```python
result['date_local'] = result['date_local'].astype(str)
result['count'] = 1

result['year']= result['date_local'].str[:4]
#region
region = result.groupby(['year','aqi_category'],as_index = False)['count'].sum()
region_final = region.pivot(index='year', columns='aqi_category', values='count')
region_final.reset_index(inplace=True)
#print(region_final.columns)
if 'Unhealthy for Sensitive Groups' not in region_final.columns:
    region_final['Unhealthy for Sensitive Groups']=0
if 'Unhealthy' not in region_final.columns:
    region_final['Unhealthy'] = 0
if 'Very Unhealthy' not in region_final.columns:
    region_final['Very Unhealthy'] = 0
region_final = region_final[['year', 'Unhealthy for Sensitive Groups','Unhealthy','Very Unhealthy']]
region_final
```

```python
#county
county = result.groupby(['year','aqi_category','id'],as_index = False)['count'].sum()
pivot_df_county = county.pivot_table(index=['id','year'], columns='aqi_category', values='count')
county_final = pivot_df_county.fillna(0)
county_final.reset_index(inplace=True)
if 'Unhealthy for Sensitive Groups' not in county_final.columns:
    county_final['Unhealthy for Sensitive Groups']=0
if 'Unhealthy' not in county_final.columns:
    county_final['Unhealthy'] = 0
if 'Very Unhealthy' not in county_final.columns:
    county_final['Very Unhealthy'] = 0
county_final = county_final[['id','year', 'Unhealthy for Sensitive Groups','Unhealthy','Very Unhealthy']]
county_final
```

<h2>Organize into function</h2>

```python
def indicator2_1b(df):
    df_select = df.loc[:,['state_code','county_code','first_max_value','date_local']]
    df_select['state_code'] = df_select['state_code'].astype('str')
    df_select['county_code']= df_select['county_code'].astype('str')
    df_select['state_code'] = df_select['state_code'].apply(lambda x: x.zfill(2))
    df_select['county_code'] = df_select['county_code'].apply(lambda x: x.zfill(3))
    df_select['id'] = df_select['state_code']+df_select['county_code']
    df_require = df_select.loc[df_select['id'].isin(['39101','39091', '39159', '39097', '39047', '39041','39117', '39049','39129','39141','39083', '39089','39045', '39127', '39073'])]
    df_require1 = df_require.loc[df_require['first_max_value'] > 35.4]
    if len(df_require1)!= 0:
        df_require1.loc[(df_require1['first_max_value'] >35.4) & (df_require1['first_max_value'] <= 55.4), 'aqi_category'] = "Unhealthy for Sensitive Groups"
        df_require1.loc[(df_require1['first_max_value'] >55.4) & (df_require1['first_max_value'] <= 150.4), 'aqi_category'] = "Unhealthy"
        df_require1.loc[df_require1['first_max_value'] >150.4 , 'aqi_category'] = "Very Unhealthy"
        result = df_require1.drop_duplicates(subset=['id','date_local'], keep="last")
        result['date_local'] = result['date_local'].astype(str)
        result['count'] = 1
        result['year']= result['date_local'].str[:4]
        #region
        region = result.groupby(['year','aqi_category'],as_index = False)['count'].sum()
        region_final = region.pivot(index='year', columns='aqi_category', values='count')
        region_final.reset_index(inplace=True)
    #print(region_final.columns)
        if 'Unhealthy for Sensitive Groups' not in region_final.columns:
            region_final['Unhealthy for Sensitive Groups']=0
        if 'Unhealthy' not in region_final.columns:
            region_final['Unhealthy'] = 0
        if 'Very Unhealthy' not in region_final.columns:
            region_final['Very Unhealthy'] = 0
        region_final = region_final[['year', 'Unhealthy for Sensitive Groups','Unhealthy','Very Unhealthy']]
        #county
        county = result.groupby(['year','aqi_category','id'],as_index = False)['count'].sum()
        pivot_df_county = county.pivot_table(index=['id','year'], columns='aqi_category', values='count')
        county_final = pivot_df_county.fillna(0)
        county_final.reset_index(inplace=True)
        if 'Unhealthy for Sensitive Groups' not in county_final.columns:
            county_final['Unhealthy for Sensitive Groups']=0
        if 'Unhealthy' not in county_final.columns:
            county_final['Unhealthy'] = 0
        if 'Very Unhealthy' not in county_final.columns:
            county_final['Very Unhealthy'] = 0
        county_final = county_final[['id','year', 'Unhealthy for Sensitive Groups','Unhealthy','Very Unhealthy']]
    else:
        region_final = pd.DataFrame(columns = ['year', 'Unhealthy for Sensitive Groups','Unhealthy','Very Unhealthy'])
        county_final = pd.DataFrame(columns = ['id','year', 'Unhealthy for Sensitive Groups','Unhealthy','Very Unhealthy'])
    return region_final,county_final
```

```python
y_list = [x for x in range(2009,2020)]
def get_data(year):
    url = "https://aqs.epa.gov/data/api/dailyData/byState?email=test@aqs.api&key=test&param=88101&bdate={}0101&edate={}1231&state=39".format(str(year),str(year))
    response = requests.get(url)
    
    dictr= response.json()
    print(dictr['Header'])
    data = dictr['Data']
    df = json_normalize(data)
    return df
```

```python
region1 = pd.DataFrame(columns = ['year', 'Unhealthy for Sensitive Groups','Unhealthy','Very Unhealthy'])
county1 = pd.DataFrame(columns = ['id','year', 'Unhealthy for Sensitive Groups','Unhealthy','Very Unhealthy'])
for i in y_list:
    df = get_data(i)
    region_final,county_final = indicator2_1b(df)
    region1 = pd.concat([region1,region_final])
    county1 = pd.concat([county1,county_final])
```
